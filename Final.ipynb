## Visualize data distributions

```python

fig = plt.figure(figsize=(25, 30))

# 1. Distribution of features
numeric_features = ['sleep_hours', 'mood_level', 'stress_level', 'screen_time', 
                   'hydration_level']

for i, feature in enumerate(numeric_features,1):
    plt.subplot(6, 3, i)
    plt.hist(data[feature], bins=30, alpha=0.7, edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
```

```python
 severity_dist = data[data['migraine_occurrence'] == 1]['migraine_severity'].value_counts().sort_index()
plt.bar(severity_dist.index, severity_dist.values)
plt.title('Migraine Severity Distribution')
plt.xlabel('Severity Level')
plt.ylabel('Count')
```
